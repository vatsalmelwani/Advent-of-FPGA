```python
def find_sum(l):
    m_outer = l[0]
    m_inner = l[1]
    for i in range(1, len(l)-1):
        if l[i] > m_outer:
            m_outer = l[i]
            m_inner = l[i+1]
            continue
        elif l[i] > m_inner:
            m_inner = l[i]
    if l[-1]>m_inner:
        m_inner = l[-1]
    # print(m_outer*10 + m_inner)
    return m_outer*10 + m_inner
```

```python
result = 0
with open('inp.txt', 'r') as f:
    for line in f:
        line = line.strip()
        nums = [int(ch) for ch in line]
        # print(nums)
        result += find_sum(nums)
        print(result)
print(result)
```

```python
#!/usr/bin/env python3
from pathlib import Path

IN  = Path("inp.txt")
OUT = Path("output.mem")

if not IN.exists():
    raise SystemExit(f"{IN} not found")

with IN.open() as fin, OUT.open("w") as fout:
    written = 0
    for raw in fin:
        s = raw.strip()
        if not s:
            continue
        # assume s is a decimal integer
        n = int(s)            # will raise if non-numeric
        fout.write(format(n, "x") + "\n")
        written += 1

print(f"Wrote {written} lines to {OUT}")
```

```python
def max_subsequence_number(num_str, k=12):
    """
    Returns the largest k-digit number (as int) formed by selecting
    digits from num_str while maintaining relative order.
    """
    stack = []
    drops = len(num_str) - k

    for ch in num_str:
        while drops > 0 and stack and stack[-1] < ch:
            stack.pop()
            drops -= 1
        stack.append(ch)

    # Take only first k digits
    return int(''.join(stack[:k]))


def sum_max_12digit_numbers_from_file(filename, count=200):
    """
    Reads `count` numbers from a text file (one per line),
    computes the largest 12-digit subsequence for each,
    and returns their sum.
    """
    total = 0

    with open(filename, 'r') as f:
        for i, line in enumerate(f):
            if i >= count:
                break
            num_str = line.strip()
            if not num_str:
                continue
            total += max_subsequence_number(num_str, 12)
            print(max_subsequence_number(num_str, 12))

    return total
```

```python
result = sum_max_12digit_numbers_from_file("inp.txt", 200)
print(result)
```
